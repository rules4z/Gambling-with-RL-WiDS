# Numpy Implementation

```python
import numpy as np
# Create a 2D Numpy array of size 1x3 with elements of your choice
arr1=np.array([[1,2,3]])

# Create a Numpy array of length 50 with zeroes as its elements
arr2=np.zeros((50))

#Create a Numpy array of length 3x2 with elements of your choice
arr3=np.array([[1,2],[3,4],[5,6]])

#Multiply arr1 and arr3 using Numpy functions
arr4=np.dot(arr1, arr3)

#Change 5th element of arr2 to a different number
arr2[4]=2

if np.shape(arr4)==(1,2) and arr2[4]!=0:
  print("Passed")
else:
  print("Fail")
```

# Matplotlib Implementation

```python
import matplotlib.pyplot as plt
import numpy as np



xpoints=np.array([1,2,3,4])
ypoints=np.array([2,4,6,8])

#Plot these points without drawing a line
plt.figure()
plt.plot(xpoints, ypoints, linestyle='', marker='.')
plt.show()

#Plotting with marker: Plot these points with a marker(Star marker)
plt.figure()
plt.plot(xpoints, ypoints, marker='*')
plt.show()

#Using fmt format, add circular marker,red color and Dashed line
plt.figure()
plt.plot(xpoints, ypoints, 'ro--')

#Add xlabel,ylabel and title for the plot.
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Graph")
plt.show()

#Create a scatter plot for xpoints and ypoints
plt.figure()
l=plt.scatter(xpoints, ypoints)

#Set color to the scatter plot. Blue,Green,Red and yellow color for each point respectively
col=['blue','green','red','yellow']
l.set_color(col)
plt.show()
```

# Miscellaneous Modules Implementation

```python
import random
import numpy as np

#Set the seed of random to 20
np.random.seed(20)

arr1=np.array([1,24,31,45,73,81,94,25])

#Using the random module pick 4 different random numbers from arr1 and return their sum.
x=np.random.choice(arr1, size=4,replace=False)
print(x)
print(np.sum(x))
```
